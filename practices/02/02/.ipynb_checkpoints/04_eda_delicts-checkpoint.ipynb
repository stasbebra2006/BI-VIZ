#### ğŸ† Typy
Sloupec *types* chceme pÅ™etypovat z Å™etÄ›zce na pole:
* rozdÄ›lÃ­me Å™etÄ›zec na pole podÅ™etÄ›zcÅ¯ podle ÄÃ¡rek
* podÅ™etÄ›zce pÅ™edstavujÃ­cÃ­ jednotlivÃ© identifikÃ¡tory pÅ™etypujeme na celÃ¡ ÄÃ­sla

```python
# splits string into list of substrings
substr = '18,19,22'.split(',')
print(substr)

# casts every element of substr to integer and returns map object
print(map(int, substr))

# casts result to list
print(list(map(int, substr)))
```

```python
delicts.types = delicts.types.apply(lambda x: list(map(int, x.split(','))))
```

#### ğŸ‘‘ HlavnÃ­ typy
PotÃ©, co jsme pÅ™etypovali sloupec *types* na pole, je nynÃ­ moÅ¾nÃ© pro kaÅ¾dÃ½ typ deliktu z tohoto pole najÃ­t jeho `parent_id`. Pro potÅ™eby vizualizacÃ­ vytvoÅ™Ã­me novÃ½ sloupec *main_types*, kterÃ½ bude obsahovat jen hlavnÃ­ typy.

```python
def get_parent_ids(ids):
    parent_ids = set()
    for type_id in ids:
        # finds parent_id for delict type with provided type_id
        parent_id = types.loc[types.id == type_id, 'parent_id'].iloc[0]
        parent_ids.add(parent_id)
    return parent_ids

delicts['main_types'] = delicts.types.apply(get_parent_ids)
```

```python
def bar_chart(x, y, xlabel, ylabel, title, rotate_labels=False):
    # fig and ax init
    fig, ax = plt.subplots(figsize=(16,8))

    # bar chart
    rects = ax.bar(x, y, color='xkcd:sky blue')

    # makes chart easier to understand
    ax.bar_label(rects, padding=3)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_title(title, fontsize=14)
    
    if rotate_labels:
        ax.set_xticks(ax.get_xticks(), x, rotation=45)

    # styling
    ax.spines['top'].set_visible(False) 
    ax.spines['right'].set_visible(False)
```

# ğŸ§ JakÃ½ je pomÄ›r validnÃ­ch a nevalidnÃ­ch evidovanÃ½ch deliktÅ¯?
V sekci na seznÃ¡menÃ­ s daty jsme si Å™Ã­kali, Å¾e delikty mÅ¯Å¾eme podle jejich stavu rozdÄ›lit na validnÃ­ (1 - *zjiÅ¡tÄ›n pravdÄ›podobnÃ½ pachatel* a 2 - *neobjasnÄ›no*) a nevalidnÃ­ (3 - *skutek nenÃ­ trestnÃ½m Äinem* a 4 - *skutek se nestal*).

```python
# delicts.state < 3 returns boolean Series
data = (delicts.state < 3).value_counts()
data
```

```python
def pie_plot(values, labels, title, ax=None):
    # fig and ax init
    if ax == None:
        fig, ax = plt.subplots()

    # pie chart
    ax.pie(values, labels=labels, autopct='%1.1f%%', colors=['xkcd:goldenrod', 'xkcd:dusty rose'])
    ax.set_title(title, fontsize=14)

    # equal aspect ratio ensures that pie is drawn as a circle
    _ = ax.axis('equal')
```

```python
labels = ['ValidnÃ­ delikty', 'NevalidnÃ­ delikty']
pie_plot(data.values, labels, 'PomÄ›r validnÃ­ch a nevalidnÃ­ch deliktÅ¯')
```

# ğŸ§ JakÃ½ je pomÄ›r evidovanÃ½ch pÅ™estupkÅ¯ a trestnÃ½ch ÄinÅ¯?
Jako pÅ™estupek budeme brÃ¡t kaÅ¾dÃ½ delikt, kterÃ½ ve sloupci main_types **obsahuje id 97**. Jako trestnÃ½ Äin budeme brÃ¡t vÅ¡echny ostatnÃ­ delikty.

```python
def get_offense_crime_cnts(df):
    # sum of boolean Series returns number of True elements 
    offense_cnt = df.main_types.apply(lambda x: 97 in x).sum()
    crime_cnt = df.shape[0] - offense_cnt    
    return [offense_cnt, crime_cnt]

def offense_crime_pie_plot(df, title='PomÄ›r pÅ™estupkÅ¯ a trestnÃ½ch ÄinÅ¯', ax=None):
    pie_plot(get_offense_crime_cnts(df), ['PÅ™estupky', 'TrestnÃ© Äiny'], title, ax=ax)
```

```python
offense_crime_pie_plot(delicts)
```

VidÃ­me, Å¾e pÅ™estupky tvoÅ™Ã­ vÄ›tÅ¡inu deliktÅ¯ â€“ aÅ¾ 84.3%. Bude se tento pomÄ›r vÃ½raznÄ› mÄ›nit, kdyÅ¾ se zamÄ›Å™Ã­me pouze na validnÃ­/nevalidnÃ­ delikty?

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,6))
fig.suptitle('ProcentuÃ¡lnÃ­ pomÄ›r pÅ™estupkÅ¯ a trestnÃ½ch ÄinÅ¯', fontsize=16)
offense_crime_pie_plot(delicts[valid_delicts], title='ValidnÃ­ delikty', ax=ax1)
offense_crime_pie_plot(delicts[invalid_delicts], title='NevalidnÃ­ delikty', ax=ax2)
```

U validnÃ­ch deliktÅ¯ se pomÄ›r tÃ©mÄ›Å™ nezmÄ›nil. To nenÃ­ pÅ™ekvapujÃ­cÃ­ vzhledem k tomu, Å¾e validnÃ­ delikty tvoÅ™Ã­ 99.4% vÅ¡ech deliktÅ¯. NevalidnÃ­ delikty majÃ­ vÃ½raznÄ› odliÅ¡nÃ½ pomÄ›r. TrestnÃ© Äiny tvoÅ™Ã­ vÃ­ce neÅ¾ polovinu nevalidnÃ­ch deliktÅ¯ â€“ 53.4%.

# ğŸ§ JakÃ½ trestnÃ½ Äin je nejÄastÄ›jÅ¡Ã­/nejmÃ©nÄ› ÄastÃ½?

```python
def type_decomposition(types, ids):
    counts = []
    for type_id in ids:
        count = types.apply(lambda x: type_id in x).sum()
        counts.append(count) 
    return counts
```

```python
crime_cnts = type_decomposition(delicts[crimes].main_types, main_types[main_types.crime].index)
crime_cnts
```

Data budeme vizualizovat pomocÃ­ sloupcovÃ©ho grafu. Pokud chceme, aby byl sloupcovÃ½ graf co nejvÃ­ce uÅ¾ivatelsky pÅ™Ã­vÄ›tivÃ½ ğŸ§‘, je dobrÃ© hodnoty seÅ™adit.

```python
# works only for arrays with non-negative values
def convert_to_percentage(arr):
    base = sum(arr)
    return (arr / base) * 100

valid_crimes_percentage = convert_to_percentage(valid_crimes_cnt)
invalid_crimes_percentage = convert_to_percentage(invalid_crimes_cnt)
```

# ğŸ§ Jak se vyvÃ­jÃ­ poÄet deliktÅ¯ v Äase?
K zodpovÄ›zenÃ­ otÃ¡zky budeme pro kaÅ¾dÃ½ den potÅ™ebovat poÄet deliktÅ¯, kterÃ© nastaly. KromÄ› toho k tomu vyuÅ¾ijeme metodu [resample](https://pandas.pydata.org/docs/reference/api/pandas.Series.resample.html), o kterÃ© si povÃ­me vÃ­ce pÅ™i cviÄenÃ­ na ÄasovÃ© Å™ady.

```python
# returns number of delicts for every hour of every day from min to max date in Series
delicts.date.value_counts()
```

```python
daily_counts = delicts.date.value_counts().resample('d').sum()
daily_counts
```

```python
daily_counts_avg = daily_counts.rolling(window=7).mean()
daily_counts_avg
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(daily_counts.index, daily_counts.values, 'o-', color='xkcd:muted purple', label='PoÄet deliktÅ¯')
ax.plot(daily_counts_avg.index, daily_counts_avg.values, '-', color='xkcd:faded orange', label='KlouzavÃ½ prÅ¯mÄ›r (7 dnÃ­)')

ax.grid(axis='y', color='black', alpha=.2, linewidth=.5)
ax.set_xlabel('Den')

ax.set_title('ÄŒasovÃ½ vÃ½voj poÄtu deliktÅ¯ za den', fontsize=14)
_ = ax.legend()
```

Na grafu jsme zobrazili poÄet evidovanÃ½ch deliktÅ¯ ke kaÅ¾dÃ©mu dni a takÃ© 7-dennÃ­ klouzavÃ½ mediÃ¡n (prÅ¯mÄ›r z pÅ™edchozÃ­ch 7 dnÅ¯). Za uvaÅ¾ovanÃ© obdobÃ­ prÅ¯mÄ›rnÃ½ poÄet deliktÅ¯ vzrostl.
